```python
# Símbolos y parámetro (suponemos a>0 para evitar |a| en normas)
u = sp.symbols('u', real=True)
a = sp.symbols('a', positive=True, real=True)

# Curva r(u)
x = a*u*(3 - u**2)
y = 3*a*u**2
z = a*u*(3 + u**2)
r = sp.Matrix([x, y, z])

# r'(u) y ds/du = ||r'(u)||
rp = sp.diff(r, u)
norm_rp = sp.sqrt((rp.T*rp)[0])
ds_du = sp.simplify(norm_rp)                     # 3*sqrt(2)*a*(1+u**2)

# --- Paso 1: u del origen ---
u_origen = sp.Integer(0)

# --- Paso 2: u del punto (2a,3a,4a) ---
cands = sp.solve(sp.Eq(y, 3*a), u)               # y=3a -> u=±1
# verificamos x y z
u_goal = [uu for uu in cands
          if sp.simplify(x.subs(u,uu)-2*a)==0 and sp.simplify(z.subs(u,uu)-4*a)==0][0]

# --- Paso 3: Longitud L = ∫_0^{u_goal} ds/du du ---
L_val = sp.simplify(sp.integrate(ds_du, (u, u_origen, u_goal)))  # = 4*sqrt(2)*a

# --- Salida bonita (MathJax) ---
display(Math(r"\textbf{Inciso 9(b): Longitud desde el origen hasta }(2a,3a,4a)"))
display(Math(r"\mathbf r(u)=\big(a\,u(3-u^2),\ 3a\,u^2,\ a\,u(3+u^2)\big)"))
display(Math(r"\frac{ds}{du}=\lVert\mathbf r'(u)\rVert = "+ sp.latex(ds_du)))
display(Math(r"\text{Origen: }u=0,\quad \text{y=3a} \Rightarrow u=\pm 1,\ "
             r"\text{verificando }x=2a,\ z=4a\Rightarrow u=1."))
display(Math(r"L=\int_{0}^{1}\frac{ds}{du}\,du"
             r"=\int_{0}^{1}"+sp.latex(ds_du)+r"\,du"
             r"= "+ sp.latex(L_val)))
display(Math(r"\boxed{\,L=4\sqrt{2}\,a\,}"))




```

```python
# Inciso 9(c) en Colab: Radio de curvatura rho(u)
import sympy as sp
from IPython.display import Math, display
sp.init_printing(use_latex='mathjax')

# Símbolos (suponemos a>0 para evitar |a| en normas)
u = sp.symbols('u', real=True)
a = sp.symbols('a', positive=True, real=True)

# Curva r(u)
x = a*u*(3 - u**2)
y = 3*a*u**2
z = a*u*(3 + u**2)
r = sp.Matrix([x, y, z])

# Derivadas
rp  = sp.diff(r, u)        # r'(u)
rpp = sp.diff(rp, u)       # r''(u)

# Normas auxiliares
norm = lambda v: sp.sqrt((v.T*v)[0])
norm_rp  = sp.simplify(norm(rp))               # ||r'||
cross    = sp.simplify(rp.cross(rpp))          # r' x r''
norm_cr  = sp.simplify(norm(cross))            # ||r' x r''||

# Curvatura y radio
kappa = sp.simplify(norm_cr / norm_rp**3)
rho   = sp.simplify(1/kappa)

# Valor en u=1 (punto (2a,3a,4a))
kappa_1 = sp.simplify(kappa.subs(u, 1))
rho_1   = sp.simplify(rho.subs(u, 1))

# Salida bonita (MathJax)
display(Math(r"\textbf{Inciso 9(c): Radio de curvatura } \rho(u)"))
display(Math(r"\mathbf r(u)=\big(a\,u(3-u^2),\ 3a\,u^2,\ a\,u(3+u^2)\big)"))

display(Math(r"\mathbf r'(u)= "+ sp.latex(rp)))
display(Math(r"\mathbf r''(u)= "+ sp.latex(rpp)))

display(Math(r"\mathbf r'(u)\times \mathbf r''(u)= "+ sp.latex(cross)))
display(Math(r"\left\lVert \mathbf r'\times \mathbf r'' \right\rVert = "+ sp.latex(norm_cr)))
display(Math(r"\left\lVert \mathbf r' \right\rVert^{3} = "+ sp.latex(sp.simplify(norm_rp**3))))

display(Math(r"\kappa(u)=\dfrac{\left\lVert \mathbf r'\times \mathbf r'' \right\rVert}{\left\lVert \mathbf r' \right\rVert^{3}}"
             r"= "+ sp.latex(kappa)))
display(Math(r"\rho(u)=\dfrac{1}{\kappa(u)}= "+ sp.latex(rho)))

display(Math(r"\text{En }u=1:\quad \kappa(1)="+ sp.latex(kappa_1) + r",\quad \rho(1)="+ sp.latex(rho_1)))
```

```python
# Inciso 9(d) en Colab: Torsión τ(u)

# Símbolos (asumimos a>0 para evitar |a| en normas)
u = sp.symbols('u', real=True)
a = sp.symbols('a', positive=True, real=True)

# Curva r(u)
x = a*u*(3 - u**2)
y = 3*a*u**2
z = a*u*(3 + u**2)
r = sp.Matrix([x, y, z])

# Derivadas
rp   = sp.diff(r, u)        # r'(u)
rpp  = sp.diff(rp, u)       # r''(u)
rppp = sp.diff(rpp, u)      # r'''(u)

# Normas auxiliares
norm = lambda v: sp.sqrt((v.T*v)[0])
cross = sp.simplify(rp.cross(rpp))             # r' × r''
norm_cross_sq = sp.simplify(norm(cross)**2)    # ||r'×r''||^2

# Torsión τ(u) = det(r', r'', r''') / ||r'×r''||^2
det_triple = sp.simplify(sp.det(sp.Matrix.hstack(rp, rpp, rppp)))
tau = sp.simplify(det_triple / norm_cross_sq)

# Valor en u=1 (punto (2a,3a,4a))
tau_1 = sp.simplify(tau.subs(u, 1))

# --- Salida bonita (MathJax) ---
display(Math(r"\textbf{Inciso 9(d): Torsión } \tau(u)"))
display(Math(r"\mathbf r(u)=\big(a\,u(3-u^2),\ 3a\,u^2,\ a\,u(3+u^2)\big)"))

display(Math(r"\mathbf r'(u)= "+ sp.latex(rp)))
display(Math(r"\mathbf r''(u)= "+ sp.latex(rpp)))
display(Math(r"\mathbf r'''(u)= "+ sp.latex(rppp)))

display(Math(r"\mathbf r'(u)\times \mathbf r''(u)= "+ sp.latex(cross)))
display(Math(r"\left\lVert \mathbf r' \times \mathbf r'' \right\rVert^{2}= "+ sp.latex(norm_cross_sq)))

display(Math(r"\det(\mathbf r',\mathbf r'',\mathbf r''') = (\mathbf r'\times\mathbf r'')\cdot \mathbf r''' = "
             + sp.latex(det_triple)))

display(Math(r"\tau(u)=\dfrac{\det(\mathbf r',\mathbf r'',\mathbf r''')}{\left\lVert \mathbf r'\times \mathbf r'' \right\rVert^{2}}"
             r"= "+ sp.latex(tau)))

display(Math(r"\text{En }u=1:\quad \tau(1)= "+ sp.latex(tau_1)))
```

```python
# Inciso 9(e) en Colab: Triedro de Frenet–Serret (T, N, B) y verificación

# Símbolos (asumimos a>0 para evitar |a| en normas)
u = sp.symbols('u', real=True)
a = sp.symbols('a', positive=True, real=True)

# Curva r(u)
x = a*u*(3 - u**2)
y = 3*a*u**2
z = a*u*(3 + u**2)
r = sp.Matrix([x, y, z])

# Derivadas
rp  = sp.diff(r, u)         # r'(u)
rpp = sp.diff(rp, u)        # r''(u)
rppp= sp.diff(rpp, u)       # r'''(u)

# Utilidades
norm = lambda v: sp.sqrt((v.T*v)[0])

# Magnitudes básicas
nrp   = sp.simplify(norm(rp))            # ||r'||
cross = sp.simplify(rp.cross(rpp))       # r' × r''
ncross= sp.simplify(norm(cross))         # ||r' × r''||

# Curvatura y torsión (para las ecuaciones de F-S)
kappa = sp.simplify(ncross / nrp**3)
tau   = sp.simplify(sp.det(sp.Matrix.hstack(rp, rpp, rppp)) / (ncross**2))

# Triedro de Frenet–Serret
T = sp.simplify(rp / nrp)
B = sp.simplify(cross / ncross)
N = sp.simplify(B.cross(T))  # N = B × T

# Salida bonita
display(Math(r"\textbf{Inciso 9(e): Triedro de Frenet–Serret}"))
display(Math(r"\mathbf r(u)=\big(a\,u(3-u^2),\ 3a\,u^2,\ a\,u(3+u^2)\big)"))

display(Math(r"\mathbf T(u)= "+ sp.latex(T)))
display(Math(r"\mathbf B(u)= "+ sp.latex(B)))
display(Math(r"\mathbf N(u)=\mathbf B\times\mathbf T= "+ sp.latex(N)))

# Ortonormalidad
display(Math(r"\text{Ortonormalidad: } \|\mathbf T\|^2,\ \|\mathbf N\|^2,\ \|\mathbf B\|^2,\ "
             r"\mathbf T\!\cdot\!\mathbf N,\ \mathbf T\!\cdot\!\mathbf B,\ \mathbf N\!\cdot\!\mathbf B"))
display(Math(sp.latex(sp.simplify(T.dot(T)))))
display(Math(sp.latex(sp.simplify(N.dot(N)))))
display(Math(sp.latex(sp.simplify(B.dot(B)))))
display(Math(sp.latex(sp.simplify(T.dot(N)))))
display(Math(sp.latex(sp.simplify(T.dot(B)))))
display(Math(sp.latex(sp.simplify(N.dot(B)))))

# Mano derecha: T × N = B
display(Math(r"\text{Mano derecha: } \mathbf T\times\mathbf N-\mathbf B = \mathbf 0"))
display(Math(sp.latex(sp.simplify(T.cross(N) - B))))

# Ecuaciones de Frenet–Serret (respecto a s): d/ds = (1/||r'||) d/du
dT_du = sp.diff(T, u); dN_du = sp.diff(N, u); dB_du = sp.diff(B, u)
dT_ds = sp.simplify(dT_du / nrp)
dN_ds = sp.simplify(dN_du / nrp)
dB_ds = sp.simplify(dB_du / nrp)

display(Math(r"\textbf{Ecuaciones de Frenet–Serret (verificación)}}"))
display(Math(r"\frac{d\mathbf T}{ds}-\kappa\,\mathbf N = "))
display(Math(sp.latex(sp.simplify(dT_ds - kappa*N))))
display(Math(r"\frac{d\mathbf N}{ds}+ \kappa\,\mathbf T - \tau\,\mathbf B = "))
display(Math(sp.latex(sp.simplify(dN_ds + kappa*T - tau*B))))
display(Math(r"\frac{d\mathbf B}{ds}+ \tau\,\mathbf N = "))
display(Math(sp.latex(sp.simplify(dB_ds + tau*N))))

# Triedro en el punto (2a,3a,4a), que corresponde a u=1
display(Math(r"\textbf{Valores en } u=1 \text{ (punto }(2a,3a,4a)\text{)}"))
display(Math(r"\mathbf T(1)= "+ sp.latex(sp.simplify(T.subs(u,1)))))
display(Math(r"\mathbf N(1)= "+ sp.latex(sp.simplify(N.subs(u,1)))))
display(Math(r"\mathbf B(1)= "+ sp.latex(sp.simplify(B.subs(u,1)))))
display(Math(r"\kappa(1)= "+ sp.latex(sp.simplify(kappa.subs(u,1))) + r",\quad "
             r"\tau(1)= "+ sp.latex(sp.simplify(tau.subs(u,1)))))
```
